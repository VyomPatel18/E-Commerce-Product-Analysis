Create Date Features

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
df
```

```python
# Feature Engineering
df['year'] = df['Purchase_Date'].dt.year
df['month'] = df['Purchase_Date'].dt.month
df['day'] = df['Purchase_Date'].dt.day
df['day_of_week'] = df['Purchase_Date'].dt.dayofweek  # Monday=0, Sunday=6
df['week_day_name'] = df['Purchase_Date'].dt.day_name()
df['week_number'] = df['Purchase_Date'].dt.isocalendar().week  # ISO week number
df['is_weekend'] = df['day_of_week'].isin([5, 6])     # Saturday=5, Sunday=6 -> Boolean
```

```python
df
```

```python
df.columns
```

Calculate Sales Trend

```python
total_sales = df['Final_Price(Rs.)'].sum()
print("Total Sales:", total_sales)
```

```python
sales_by_category = df.groupby('Category')['Final_Price(Rs.)'].sum()
print('Total Sales by Category:', sales_by_category)
```

```python
sales_by_paymentmethod = df.groupby('Payment_Method')['Final_Price(Rs.)'].sum()
print('Total Sales by Paymentmethod:', sales_by_paymentmethod)
```

```python
daily_sales = df.groupby(['month','day'])['Final_Price(Rs.)'].sum().reset_index()
daily_sales.columns = ['month','day','total_sales']
```

```python
daily_sales
```

```python
monthly_sales = df.groupby(['year', 'month'])['Final_Price(Rs.)'].sum().reset_index()
monthly_sales['Month_Year'] = pd.to_datetime(monthly_sales[['year', 'month']].assign(DAY=1))
monthly_sales = monthly_sales[['Month_Year', 'Final_Price(Rs.)']]
monthly_sales.columns = ['month', 'Total_Sales']
```
